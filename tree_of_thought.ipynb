## From Linear to Lateral: The Tree-of-Thoughts (ToT) Advantage

Standard AI reasoning often follows a single, fragile path. If the first step is wrong, the entire conclusion collapses. The **Tree-of-Thoughts (ToT)** framework, implemented in `tot.py`, transforms this linear process into a dynamic search tree.

### The Core Mechanism
Instead of one "leap" of logic, the code decomposes problems into discrete **thought units**:
* **Branching:** At each stage, the LLM generates multiple potential next steps (branches) rather than just one.
* **Evaluation:** A "Critic" logic scores each branch for promise (0.0 to 1.0). This identifies which paths are likely to lead to a solution and which are dead ends.
* **Pruning & Search:** Using strategies like **BFS** or **Beam Search**, the system discards low-scoring branches (pruning) and focuses hardware resources only on the most "promising" thoughts.

### Why It Matters
This approach allows the AI to **backtrack**—if a path becomes illogical, it returns to a previous "winning" state and tries a different direction. By mimicking human trial-and-error, ToT significantly reduces hallucinations and excels at complex puzzles, math, and strategic planning where a single mistake is fatal.

```python
from clients.ollama import OllamaClient
from rag.tot import TreeOfThoughtsReasoner # Save the code above as tot.py

client = OllamaClient()
reasoner = TreeOfThoughtsReasoner(client, max_depth=3, branch_factor=3)

problem = "If I have 3 apples and buy 2 more, then give 1 to a friend, how many do I have?"
final_answer = reasoner.solve(problem)
print(f"\n🎯 FINAL ANSWER: {final_answer}")
```
