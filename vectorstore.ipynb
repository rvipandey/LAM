```python
from clients.ollama import OllamaClient,OllamaConfig
from vectormemory.vector import VectorMemory,TreeOfThoughtsReasoner
config = OllamaConfig(max_tokens=2048, temperature=0.1)
client = OllamaClient(config)
```

```python
import time
memory = VectorMemory()
tot = TreeOfThoughtsReasoner(client, max_depth=2, branch_factor=2)

# 3. Index Knowledge (RAG)
knowledge_base = [
    ("k1", "Python is an interpreted, high-level programming language known for readability."),
    ("k2", "Vector databases allow finding semantically similar items using high-dimensional math."),
    ("k3", "Tree of Thoughts is a framework that allows LLMs to explore multiple reasoning paths."),
    ("k4", "Ollama enables running LLMs like Llama3 locally on your machine."),
]

print("📚 Indexing Knowledge Base...")
for uid, text in knowledge_base:
    memory.add(uid, text)

# 4. User Query
query = "How can I use Python to understand LLM reasoning?"
print(f"\n❓ Query: {query}")

# 5. Retrieve Context (Fast, local calculation)
print("🔍 Retrieving relevant context...")
context_docs = memory.retrieve(query, k=2)
context_str = "\n".join([f"- {doc}" for doc in context_docs])
print(f"📄 Found Context:\n{context_str}")

# 6. Reason with ToT (Optimized)
print("\n🧠 Reasoning with Tree of Thoughts...")
# Inject the RAG context into the ToT problem statement
augmented_problem = f"User Question: {query}\n\nRelevant Context:\n{context_str}\n\nProvide a detailed answer."

start_time = time.time()
final_answer = tot.solve(augmented_problem)
end_time = time.time()

print("\n✅ Final Answer:")
print("-" * 50)
print(final_answer)
print("-" * 50)
print(f"\n⏱️ Total Time: {end_time - start_time:.2f} seconds")
```
